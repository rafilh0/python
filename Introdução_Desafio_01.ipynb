````
import re

def verifica_inicio(string):
    padrao = r'^[\d\W_]'  # Padrão para verificar se começa com número ou caracteres especiais
    if re.match(padrao, string):
        return True
    else:
        return False

def verifica_acentuacao_espaco(string):
    padrao = r'[\sáàãâéèêíìîóòõôúùûÁÀÃÂÉÈÊÍÌÎÓÒÕÔÚÙÛçÇ]'
    if re.search(padrao, string):
        return True
    else:
        return False

def verifica_contem_nome(string, lista):
    for item in lista:
        if item in string:
            return True
    return False

while True:
  nome_completo = input("Informe seu nome completo: ").strip()
  if len(nome_completo.split()) < 2:
    print("Favor informar, pelo menos, 1 sobrenome")
    continue
  break

lista_nomes = nome_completo.lower().split()

while True:
  usuario = input("Informe um nome de usuario: ").strip()
  if verifica_inicio(usuario):
    print("O nome do usuario não pode iniciar com um número ou um caractere especial")
    continue
  if verifica_acentuacao_espaco(usuario):
    print("O nome do usuario não pode conter acentuação ou espaço em branco")
    continue
  if not verifica_contem_nome(usuario, lista_nomes):
    print("O nome do usuário deve conter pelo menos o nome ou um dos sobrenomes")
    continue
  break


while True:
  senha = input("Informe uma senha: ").strip()
  if len(senha) < 6:
    print("A senha precisa ter, no mínimo, 6 caracteres")
    continue
  if not re.search(r'[a-z]', senha) or not re.search(r'[A-Z]', senha):
    print("A senha precisa ter, pelo menos, uma letra minúscula e uma minúscula")
    continue
  if not re.search(r'\d', senha):
    print("A senha precisa ter, pelo menos, um número")
    continue
  if not re.search(r'[\W_]', senha):
    print("A senha precisa ter, pelo menos, um caractere especial")
    continue
  if ' ' in senha:
    print("A senha não pode ter espaços em branco")
    continue
  if verifica_contem_nome(senha, lista_nomes):
    print("A senha não pode ter o nome ou um dos sobrenomes do usuário")
    continue
  break

print (f"\nTodas as entradas ok.\nNome Completo: {nome_completo}")
print (f"Usuário: {usuario}")
print (f"Senha: {senha}")






````
