### 11. Compare o volume de vendas de hidratante e creme facial com um gráfico de barras.

O gráfico de barras deve exibir o número de unidades vendidas por mês para cada produto. Adicione uma barra separada para cada produto no mesmo gráfico.

Neste mesmo gráfico, apresente o lucro total de cada mês. Como a escala de quantidade de produtos é diferente da escala de lucro, crie um eixo gêmeo (pesquise por twin axes).




O gráico deve ser semelhante a esse:

image.png

```
fig, ax1 = plt.subplots()

x = df_trad['Mês']
a = df_trad['Hidratante']
b = df_trad['Creme Facial']
c = df_trad['total_profit']

ax2 = ax1.twinx()
ax1.bar(x - 0.2, a, width=0.4, label='Hidratante', align='center')
ax1.bar(x + 0.2, b,  width=0.4, label='Vendas Creme Facial', align='center')

ax2.plot(x, c,  marker = 'o', color='green',
         markersize=7, linewidth=3,
         markerfacecolor='white',
         markeredgecolor='green',
         markeredgewidth=1)
ax1.set_xlabel('Mês')
ax1.set_ylabel('Unidades Vendidas')
ax2.set_ylabel('Lucro Total')
ax1.set_title('Vendas de Hidratante e Creme Facial')

ax1.set_xticks(x)

ax1.legend(loc='upper left');
ax1.grid(False)
```

### 12. Crie uma grade de subplots e, para cada painel, exiba um gráfico de dispersão relacionando os dados de vendas de cada produto e o total de lucro obtido no mês. Na última linha da grade deve ser apresentado um único painel com um gráfico de caixas exibindo a distribuição de vendas de cada produto.


image.png

```
fig = plt.figure(figsize=(8, 6))
grid = plt.GridSpec(3, 3,wspace=0.5, hspace=0.6)
colunas_produtos = list(df_trad.columns[1:7].values)
colors = ['orange', 'blue', 'black', 'green', 'red', 'purple']
x = df_trad['total_profit']
i=0

mpl.rcParams['axes.titlesize'] = 'small'
mpl.rcParams['xtick.labelsize'] = 8
mpl.rcParams['ytick.labelsize'] = 8

for y in range(2):
  for z in range(3):
    ax = fig.add_subplot(grid[y:y+1, z:z+1])
    ax.scatter(x,df_trad[colunas_produtos[i]],s=20,color=colors[i])
    ax.set_title(colunas_produtos[i])
    i += 1


ax = plt.subplot(grid[2, :])
ax.boxplot(df_trad.iloc[:, 1:7], labels=colunas_produtos, flierprops=dict(markersize=5))
ax.set_title("Gráficos de Caixas");
```
