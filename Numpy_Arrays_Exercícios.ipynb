8 - Crie uma matriz 10x10 com valores aleatórios e encontre os valores mínimo, máximo, média, moda e desvio padrão.

```
np.random.seed(1)
a = np.random.randint(50, size=(10,10))
print(a)
print('Mínimo: {}'.format(a.min()))
print('Máximo: {}'.format(a.max()))
print('Média: {}'.format(a.mean()))
print('Moda: {}'.format(np.argmax(np.bincount(a.flatten().astype(int)))))
print('Desvio Padrao: {}'.format(np.std(a)))
```

9 - Dada uma matriz 1D, multiplique por 10 todos os elementos que estão entre 3 e 8.

```
a = np.arange(10)
a[(a >= 3) & (a <= 8)]*10
```

```
a = np.arange(10)

# Encontre os índices dos elementos que estão entre 3 e 8
indices = np.where((a >= 3) & (a <= 8))

# Multiplique os elementos nessas posições por 10
a[indices] *= 10

# Imprima a matriz resultante
a
```

14 - Crie uma matriz 3x3 e exiba o índice e o valor de cada posição. (dica: consulte as funções np.ndenumerate e np.ndindex)

```
np.random.seed(1)
a = np.random.randint(99, size=(3,3))
print(a)
for indice, valor in np.ndenumerate(a):
    print(f"Posição {indice}: {valor}")
```

15- Como ordenar uma matriz pela enésima coluna?

```
# Opcao 1 - Extraindo e ordenando coluna
np.random.seed(1)
a = np.random.randint(99, size=(3,3))
print(f"Matriz:\n{a}")

coluna_ordenar = 1
vetor_ordenado_enesima_coluna = np.sort(a[:, coluna_ordenar])
a[:, coluna_ordenar] = vetor_ordenado_enesima_coluna

print(f"\nMatriz Ordenada pela coluna {coluna_ordenar}:\n{a}")
```

```
# Opcao 2 - Ordenando nova matriz inteira
np.random.seed(1)
a = np.random.randint(99, size=(3,3))
print(f"Matriz:\n{a}")

a_ordenada_todas_colunas = np.sort(a, axis= 0)
enesima_coluna = 1
a[:, enesima_coluna] = a_ordenada_todas_colunas[:, enesima_coluna]

print(f"\nMatriz Ordenada pela coluna {enesima_coluna}:\n{a}")
```
