1.   Qual o tamanho dessa string?
2.   Inverta a última palavra da string.

```python
str = "Caminhando contra o vento\n Sem lenço, sem documento"

#1
print("1 - Tamanho da string:",len (str))

#2
nova_str = str[-1:-10:-1]
print("\n2 - A última palavra invertida é:", nova_str)

#3
str1, str2 = str.split("\n ")
print("\n3 - Variavel1:", str1)
print("    Variavel2:", str2)

#4
import re
str3 = re.sub(r'[^\w\s]', '', str)
print("\n4 - String sem sinais de pontuação:\n", str3)
```

15. Faça uma calculadora capaz de realizar as 4 operações básicas. O programa deve receber como entrada dois números e a operação que se deseja realizar.

```python
print("15- Calculadora Básica\nEsta calculadora pode efetuar as 4 operações básicas.\n")
resultado = None
vlr=[]
n = 0

while n < 2:
  valor = input(f"Informe o número n{n+1}: ").replace(",", ".").strip()
  if not valor.replace(".", "").isdigit():
    print("Valor informado não é um número")
    continue
  vlr += [float(valor)]
  n += 1
  print("Ok\n")

while True:
  operacao = input("Informe a operação (+ - * /): ").strip()
  if operacao == "+":
    resultado = vlr[0] + vlr[1]
  elif operacao == "-":
    resultado = vlr[0] - vlr[1]
  elif operacao == "*":
    resultado = vlr[0] * vlr[1]
  elif operacao == "/":
    if vlr[1] == 0:
      print ("Erro: Divisão por zero!")
    else:
      resultado = vlr[0] / vlr[1]
  else:
    print("Valor informado para a operação inválido!")
    continue
  break

if resultado is not None:
    print(f"\nO resultado da operação {vlr[0]} {operacao} {vlr[1]} é: {resultado}")
```

22. Receba dois valores do usuário e realize a divisão. Lembre-se de tratar possíveis entradas inválidas e divisão por zero.

24. Dada a lista [2, 8, 3, 13, 7, 24, 'str', [1, 2, 11], 8, True], exiba todos os números primos existentes em seu conteúdo em ordem crescente (incluindo a lista interna).

> DICA: resolver com recursão (pesquisar)




```python
lista_dada = [2, 8, 3, 13, 7, 24, 'str', [1, 2, 11], 8, True]
lista_primos=[]

def e_primo(numero):
    if numero <= 1:
      return False
    for i in range(2, numero):
      if numero % i == 0:
        return False
    return True

def retorna_lista_primos_crescente(lista):
  global lista_primos
  for item in lista:
    if isinstance(item, list):
      retorna_lista_primos_crescente(item)
    else:
      if isinstance(item, int) and e_primo(item) and item not in lista_primos:
        lista_primos += [item]
  lista_primos.sort()
  return lista_primos

print(f"24-\nOs números primos existentes em ordem crescente são:\n{retorna_lista_primos_crescente(lista_dada)}")

```
